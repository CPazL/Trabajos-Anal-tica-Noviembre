```python
#para calcular al oponente más goleado, primero dividimos la columna Resultado en dos. Resultado de equipo local y resultado equipo oponente
df_2["Resultado Local"] = 0
df_2["Resultado Oponente"] = 0
```

```python
def separador_resultados_loc(locacion, resultado):
  if resultado is pd.NA or locacion is pd.NA:
    return None
  else:
    resultado = str(resultado)
    res_split = resultado.split(sep=":")    
  if locacion == "Home":
    return res_split[0]
  else:
    return 0
```

```python
for i in range(df_2.shape[0]):
  df_2["Resultado Local"][i]= separador_resultados_loc(df_2["Locacion"][i], df_2["Resultado"][i])
```

```python
def separador_resultados_opo(locacion, resultado):
  if resultado is pd.NA or locacion is pd.NA:
    return None
  else:
    resultado = str(resultado)
    res_split = resultado.split(sep=":")    
  if locacion == "Away":
    return res_split[1]
  else:
    return 0
```

```python
for i in range(df_2.shape[0]):
  df_2["Resultado Oponente"][i]= separador_resultados_opo(df_2["Locacion"][i], df_2["Resultado"][i])
```

```python
df_2.drop(["Resultado"], axis=1, inplace = True)
```

```python
#Una vez conseguida la lista de resultados, utilizamos un groupby y la funcion max para seleccionar al oponente qué recibió mas goles.
mas_goleado = df_2.groupby(["Resultado Local"])["Oponente"].max().iloc[-1]
print(f"El oponente que recibió mas goles en su contra fue {mas_goleado}.")
```

```python
#Para averiguar qué equipo jugó mas fechas utilizamos un groupby y contamos la cantidad de fechas.
mas_fechas = df_2.groupby(["Oponente"]).Fecha.count()
print(f"El oponente que jugó mas fechas fue: {mas_fechas}.")
```
