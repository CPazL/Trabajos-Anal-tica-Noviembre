```python
# Agregamos 3 columnas, las de A√±o, Mes y D√≠a del partido
cris['Anio'] = cris['Fecha'].dt.year
cris['Mes'] = cris['Fecha'].dt.month
cris['Dia'] = cris['Fecha'].dt.day
```

```python
# Vamos a poner Gol_favor_T y Gol_contra_T --> Para saber como termino el partido con los goles a favor y en contra
# Antes de eso me gustar√≠a ver que valores tiene la columna Resultado y vemos que siempre est√° dividido por : por lo que podemos splitear ah√≠
cris.Resultado.unique()
```

```python
# Tambi√©n tendr√≠amos que ver que valores tiene la columna Evento y vemos que solo tiene local y visitante, por lo que podr√≠amos avanzar con el split
cris.Evento.unique()
```

```python
# Hacemos una funci√≥n que luego tiraremos a todo el DataFrame para los GOLES A FAVOR
def Apertura_Resultado(evento, resultado): #Funci√≥n de Apertura, hay que darle dos par√°metros, uno con el evento (Local o Visitante) y otra con el resultado.
  res_sep= resultado.split(sep=':') #Lo splitea seg√∫n : y deja en la posici√≥n 0 el local y 1 el visitante

  if evento == 'Home': # Si es local, nos interesa la posici√≥n 0
    return res_sep[0]
  elif evento =='Away': # Si es visitante, nos interesa la posici√≥n 1
    return res_sep[1]
  else:
    return None
```

```python
# Aplicamos la funci√≥n a a columna resultado y queremos que nos complete la columna Gol_favor_T con los goles a favor
cris["Gol_favor_t"] = list(map(lambda x,y: Apertura_Resultado(x,y), cris["Evento"], cris["Resultado"]))
```

```python
# Ahora hacemos lo mismo pero con la columna "Momento_anotaci√≥n"
cris["Gol_favor_m"] = list(map(lambda x,y: Apertura_Resultado(x,y), cris["Evento"], cris["Momento_anotacion"]))
```

```python
# Hacemos una funci√≥n que luego tiraremos a todo el DataFrame para los GOLES EN CONTRA
def Apertura_Resultado_Encontra(evento, resultado): #Funci√≥n de Apertura, hay que darle dos par√°metros, uno con el evento (Local o Visitante) y otra con el resultado.
  res_sep= resultado.split(sep=':') #Lo splitea seg√∫n : y deja en la posici√≥n 0 el local y 1 el visitante

  if evento == 'Home': # Si es local, nos interesa la posici√≥n 1
    return res_sep[1]
  elif evento =='Away': # Si es visitante, nos interesa la posici√≥n 0
    return res_sep[0]
  else:
    return None
```

```python
# Aplicamos la funci√≥n a a columna resultado y queremos que nos complete la columna Gol_contra_t con los goles en contra
cris["Gol_encontra_t"] = list(map(lambda x,y: Apertura_Resultado_Encontra(x,y), cris["Evento"], cris["Resultado"]))
```

```python
# Ahora hacemos lo mismo pero con la columna "Momento_anotaci√≥n"
cris["Gol_encontra_m"] = list(map(lambda x,y: Apertura_Resultado_Encontra(x,y), cris["Evento"], cris["Momento_anotacion"]))
```

```python
# Vamos a limpiar la columna Minuto y para eso vemos que tenemos que splitear por ' y + --> Una manera es eliminar las ' y luego splitear por +
cris.Minuto.unique()
```

```python
#Eliminamos los ' y los ponemos en una nueva columna auxiliar
cris["MinutosPartidoCrudo"] = list(map(lambda x: x.replace("'",""), cris["Minuto"]))
```

```python
# Funci√≥n para sumar los minutos de partido a los minutos extras
def TotalMinutos(MinCrudo):
  res_sep= MinCrudo.split(sep='+')
  if len(res_sep) == 2:
    a = int(res_sep[0])
    b = int(res_sep[1])
  else:
    a = int(res_sep[0])
    b = 0
  return a+b
```

```python
# Corremos la funci√≥n a todos los registros
cris["MinutosPartido"] = list(map(lambda x: TotalMinutos(x), cris["MinutosPartidoCrudo"]))
```

```python
# Funci√≥n para agregar el tiempo en el que se meti√≥ el gol (Primero o Segundo Tiempo, para no perder esa info)
def Tiempo(MinCrudo):
  res_sep= MinCrudo.split(sep='+')
  if len(res_sep) == 2:
    if int(res_sep[0]) == 45:
      return "Primer Tiempo Extra"
    else:
      return "Segundo Tiempo Extra"    
  else:
    if int(res_sep[0]) <= 45:
      return "Primer Tiempo"
    else:
      return "Segundo Tiempo"    
```

```python
# Corremos la funci√≥n a todos los registros
cris["Tiempo del gol"] = list(map(lambda x: Tiempo(x), cris["MinutosPartidoCrudo"]))
```

```python
# Bajamos DataFrame y modificamos
messi= pd.read_csv("https://raw.githubusercontent.com/CPazL/clase-pandasPython/main/messiSucio.csv")

messi.rename(columns={'Tournament':'Torneo'}, inplace=True)
messi.rename(columns={'Matchday':'Jornada'}, inplace=True)
messi.rename(columns={'Date':'Fecha'}, inplace=True)
messi.rename(columns={'Venue':'Evento'}, inplace=True)
messi.rename(columns={'Opponent':'Oponente'}, inplace=True)
messi.rename(columns={'Result':'Resultado'}, inplace=True)
messi.rename(columns={'Position':'Posici√≥n'}, inplace=True)
messi.rename(columns={'Minute':'Minuto'}, inplace=True)
messi.rename(columns={'Opponent':'Oponente'}, inplace=True)
messi.rename(columns={'When Scored':'Momento de anotaci√≥n'}, inplace=True)
messi.rename(columns={'Goal Type':'Tipo_de_gol'}, inplace=True)
messi.rename(columns={'Assist':'Asistente'}, inplace=True)

messi['Torneo'] = messi['Torneo'].astype('string')
messi['Jornada'] = messi['Jornada'].astype('string')
messi['Fecha'] = pd.to_datetime(messi['Fecha'])
messi['Evento'] = messi['Evento'].astype('string')
messi['Club'] = messi['Club'].astype('string')
messi['Oponente'] = messi['Oponente'].astype('string')
messi['Resultado'] = messi['Resultado'].astype('string')
messi['Posici√≥n'] = messi['Posici√≥n'].astype('string')
messi['Tipo de gol'] = messi['Tipo_de_gol'].astype('string')
messi['Asistente'] = messi['Asistente'].astype('string')
messi['Pareja'] = messi['Pareja'].astype('string')
messi['Minuto'] = messi['Minuto'].astype('string')
messi['Momento de anotaci√≥n'] = messi['Momento de anotaci√≥n'].astype('string')

messi.dropna(axis=0, inplace= True, how='all')
messi.dropna(axis=0, inplace= True, thresh=int(messi.shape[1]*0.8))
messi.drop(['Pareja'], axis=1, inplace=True)
messi= messi.drop_duplicates()
messi['Anio'] = messi['Fecha'].dt.year

# Vamos a limpiar la columna Minuto y para eso vemos que tenemos que splitear por ' y + --> Una manera es eliminar las ' y luego splitear por +
messi.Minuto.unique()
# Eliminamos los ' y los ponemos en una nueva columna auxiliar
messi["MinutosPartidoCrudo"] = list(map(lambda x: x.replace("'",""), messi["Minuto"]))
# Corremos la funci√≥n a todos los registros
messi["MinutosPartido"] = list(map(lambda x: TotalMinutos(x), messi["MinutosPartidoCrudo"]))
# Corremos la funci√≥n a todos los registros
messi["Tiempo del gol"] = list(map(lambda x: Tiempo(x), messi["MinutosPartidoCrudo"]))
# Borramos columna auxiliar
messi.drop(["MinutosPartidoCrudo"], axis=1, inplace=True)
```

###**üßπ 1) Comparacion de goles metidos por Messi y Cristiano a√±o tras a√±o**

###**üßπ 2) Vamos a comparar en que momento del partido suelen hacer los goles cada uno**

```python
# Armo un DataFrame de Cristiano Ronaldo con la cantidad de goles que meti√≥ en cada minuto de los partidos
TorneosCris = cris.groupby("MinutosPartido").count()
TorneosCris = pd.DataFrame(TorneosCris.Club).reset_index()
TorneosCris.columns
TorneosCris = TorneosCris.rename(columns = {"Club":"Goles"})
TorneosCris
```

```python
# Armo un DataFrame de Messi con la cantidad de goles que meti√≥ en cada minuto de los partidos
TorneosMessi = messi.groupby("MinutosPartido").count()
TorneosMessi = pd.DataFrame(TorneosMessi.Club).reset_index()
TorneosMessi.columns
TorneosMessi = TorneosMessi.rename(columns = {"Club":"Goles"})
TorneosMessi
```

```python
# Hago un Merge para dejar solo un DataFrame con toda la info lista para usar el un gr√°fico

GolesPorMinuto = merged_full = pd.merge(left = TorneosCris, right = TorneosMessi, how='outer', left_on="MinutosPartido", right_on="MinutosPartido").fillna(0)
GolesPorMinuto.rename(columns={'Goles_x':'GolesCris'}, inplace=True)
GolesPorMinuto.rename(columns={'Goles_y':'GolesMessi'}, inplace=True)
GolesPorMinuto['GolesMessi'] = GolesPorMinuto['GolesMessi'].astype('int')
GolesPorMinuto['GolesCris'] = GolesPorMinuto['GolesCris'].astype('int')
GolesPorMinuto.sort_values("MinutosPartido",ascending=True,inplace=True)
GolesPorMinuto.info()
```

```python
# Vemos como queda el DF
GolesPorMinuto
```

```python
# Vamos a armar un histograma para ver en que momentos meten mas goles cada uno de los dos jugadores. Armamos funcion para dividir la x del histograma en grupos de 10 minutos:

def Corte (x):
  if x <= 10:
    return "0 a 10"
  elif x > 10 and x <= 20:
    return "11 a 20"
  elif x > 20 and x <= 30:
    return "21 a 30"
  elif x > 30 and x <= 40:
    return "31 a 40"
  elif x > 40 and x <= 50:
    return "41 a 50"
  elif x > 50 and x <= 60:
    return "51 a 60"
  elif x > 60 and x <= 70:
    return "61 a 70"
  elif x > 70 and x <= 80:
    return "71 a 80"
  elif x > 80 and x <= 90:
    return "81 a 90"
  elif x > 90 and x <= 100:
    return "91 a 100"
  elif x > 100 and x <= 110:
    return "101 a 110"
  elif x > 110 and x <= 120:
    return "111 a 120"
  else:
    return "ERROR"
```

```python
# Aplicamos la Funci√≥n a todo el DataFrame
GolesPorMinuto["CorteMinutos"] = list(map(lambda x: Corte(x), GolesPorMinuto["MinutosPartido"]))
```

```python
# Vemos como queda
GolesPorMinuto
```

```python
# Se puede ver que Messi tiene mucha m√°s estabilidad en cuanto a los momentos en los que mete goles, 
# pero por otro lado Cristiano tiene muchos m√°s goles en los primeros 10 minutos por lo que arranca muy agresivo.

MESSI = px.histogram(GolesPorMinuto,
             template = "gridon",
             x = "CorteMinutos",
             y = "GolesMessi",
             width=850, 
             height=500,
             labels = {"CorteMinutos":"Minuto Partido"},
             title = "Distribuci√≥n de goles MESSI",
             text_auto = True)
MESSI.show()

CRISTIANO = px.histogram(GolesPorMinuto,
             template = "gridon",
             x = "CorteMinutos",
             y = "GolesCris",
             width=850, 
             height=500,
             labels = {"CorteMinutos":"Minuto Partido"},
             title = "Distribuci√≥n de goles CRISTIANO",
             text_auto = True)

CRISTIANO.show()
```
