*Punto 5 Limpieza de datos*

````python
lewandowski['Tournament'] = lewandowski['Tournament'].astype('string')
lewandowski['Matchday'] = lewandowski['Matchday'].astype('string')
lewandowski['Date'] = pd.to_datetime(lewandowski['Date'])
lewandowski['Venue'] = lewandowski['Venue'].astype('string')
lewandowski['Club'] = lewandowski['Club'].astype('string')
lewandowski['Opponent'] = lewandowski['Opponent'].astype('string')
lewandowski['Result'] = lewandowski['Result'].astype('string')
lewandowski['Position'] = lewandowski['Position'].astype('string')
lewandowski['Minute'] = lewandowski['Minute'].astype('string')
lewandowski['When Scored'] = lewandowski['When Scored'].astype('string')
lewandowski['Goal Type'] = lewandowski['Goal Type'].astype('string')
lewandowski['Assist'] = lewandowski['Assist'].astype('string')
lewandowski['Nacionalidad'] = lewandowski['Nacionalidad'].astype('string')

lewandowski.info()
````

*Punto 6 Limpieza de datos*


````python
lewandowski.rename(columns={'Tournament':'Torneo'}, inplace=True)
lewandowski.rename(columns={'Matchday':'Jornada'}, inplace=True)
lewandowski.rename(columns={'Date':'Fecha'}, inplace=True)
lewandowski.rename(columns={'Venue':'Evento'}, inplace=True)
lewandowski.rename(columns={'Opponent':'Oponente'}, inplace=True)
lewandowski.rename(columns={'Result':'Resultado'}, inplace=True)
lewandowski.rename(columns={'Position':'Posición'}, inplace=True)
lewandowski.rename(columns={'Minute':'Minuto'}, inplace=True)
lewandowski.rename(columns={'When Scored':'Momento de anotación'}, inplace=True)
lewandowski.rename(columns={'Goal Type':'Tipo de gol'}, inplace=True)
lewandowski.rename(columns={'Assist':'Asistente'}, inplace=True)

lewandowski.info()
````

*Punto 7 Limpieza de datos*

> 
La línea de código está comentada ya que cuando fue eliminada por primera vez la columna luego ya no se pudo volver a ejecutar, el inplace nos asegura eso.





````python
lewandowski.drop(['Nacionalidad'], axis=1, inplace=True)

lewandowski.info()
````

*Punto 8 Limpieza de datos*

*Antes de la eliminación de duplicados.*

*Luego de la eliminación de duplicados*

````python
lewandowski = lewandowski.drop_duplicates()

lewandowski.shape
````

*Punto 9 Limpieza de datos*

````python
lewandowski.dropna(axis=0, inplace= True, how='all')
lewandowski['Año'] = lewandowski['Fecha'].dt.year

lewandowski.info()
lewandowski.head()
````

*Punto 10 Limpieza de datos*
Creo columnas con los datos "resultado izquierda" y "resultado derecha" que se usarán luego para determinar la cantidad de veces que Lewandowski abrió el marcador junto a los datos "Momento de anotación"

````python
lewandowski['rdoIzq']= 0
lewandowski['rdoDer']= 0
#lewandowski.drop(['rdoIzq','rdoDer'], axis=1, inplace=True)
lewandowski.head()
````

*Punto 1 Análisis de datos*

````python
golesHastaLaFecha = lewandowski.shape
print(f'Hasta la fecha, Lewandowski ha hecho {golesHastaLaFecha[0]} goles.')
````

*Punto 2 Análisis de datos*

````python
promGol = lewandowski.groupby(by=['Torneo']).count()
promGol = promGol.Jornada
print (f'A continuación, la cantidad que de goles de Lewandowski por torneo: {promGol}')

print (f'La cantidad de goles totales es de: {golesHastaLaFecha[0]}')
````

*Punto 3 Análisis de datos*

````python
lewandowski.head()

golesEvento = lewandowski.groupby(by=['Evento']).count()
golesEvento = golesEvento.Jornada
golesLocal = round(golesEvento[0] / golesHastaLaFecha[0] * 100,2)
golesVisitante = round(golesEvento[1] / golesHastaLaFecha[0] * 100,2)

#golesLocal
#golesVisitante

print(f'Lewandowski hizo el {golesLocal}% de los goles jugando de Local, mientras que el {golesVisitante}% los hizo jugando de visitante.')


````

*Punto 4 Análisis de datos*

````python
masGoles = lewandowski.groupby(by=['Oponente']).count()
masGoles = masGoles.Jornada
masGoles
print (f'El club al que Lewandowski le hizo más goles fue {masGoles.index.values[0]}, con un total de {masGoles[0]}.')

````

*Punto 5 Análisis de datos*

````python
masFechasOp = lewandowski.groupby(['Oponente','Fecha']).count()
masFechasOp = masFechasOp.groupby(['Oponente']).count()
masFechasOp = masFechasOp.Jornada.sort_values(ascending=False)
print(f'El oponente contra el que más partidos en los que añotó goles Lewandowski fue {masFechasOp.index.values[0]}, con un total de {masFechasOp[0]} partidos disputados.')
````

*Punto 6 Análisis de datos*

````python
lewandowski['rdoIzq']= 0
lewandowski['rdoDer']= 0

for i in lewandowski['Momento de anotación']:
  if i == '1:0':
    lewandowski.rdoIzq += 1
  elif i == '0:1':
    lewandowski.rdoDer += 1
  else:
    None  

print(f'Lewandowski abrió el marcador un total de {lewandowski.rdoIzq[0] + lewandowski.rdoDer[0]} veces')
````
